#Program to illustrate Scatter Plot

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Set plot space as inline for inline plots and qt for external plots
%matplotlib inline

# Declaring  X and y values
x =[2, 5, 6, 8, 7, 8, 6, 9,
    14, 13, 11, 19, 6] 
  
y =[79, 98, 97, 78, 90, 96, 
    103, 97, 84, 98, 87, 98, 102]

# Set the figure size in inches
plt.figure(figsize=(10,6))

plt.scatter(x, y, label = "Point", c = "blue")

# Set x and y axes labels
plt.xlabel('X Values')
plt.ylabel('Y Values')

plt.title('Scatter Plot')
plt.legend()
plt.show()
```

#Program to illustrate Box Plot

#Program to illustrate Pie Chart

```python
import matplotlib.pyplot as plt
labels = ['Python', 'C++', 'C', 'Java']
sizes = [115, 300, 145, 250]
# Plot
plt.pie(sizes, labels=labels, 
        autopct='%1.2f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```

#Program to illustrate Stacked Bar Chart

```python
# importing package
import matplotlib.pyplot as plt
import numpy as np
  
# create data
x = ['P', 'Q', 'R', 'S']
y1 = np.array([20, 25, 15, 25])
y2 = np.array([12, 15, 19, 6])
y3 = np.array([10, 29, 13, 19])
y4 = np.array([15, 25, 18, 19])
  
# plot bars in stack manner
plt.bar(x, y1, color='orange')
plt.bar(x, y2, bottom=y1, color='red')
plt.bar(x, y3, bottom=y1+y2, color='brown')
plt.bar(x, y4, bottom=y1+y2+y3, color='grey')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

#Program to illustrate Bar Plot

```python

import matplotlib.pyplot as plt  
   
# Creating the dataset 
data = {'C':15, 'C++':25, 'Java':30,  
        'Python':20} 
courses = list(data.keys()) 
values = list(data.values()) 
   

fig = plt.figure(figsize = (10, 6)) 
  
# creating the bar plot 
plt.bar(courses, values, color ='orange',  
        width = 0.4) 
  
plt.xlabel("Courses offered") 
plt.ylabel("No. of students enrolled") 
plt.title("Students enrolled in different courses") 
plt.show() 
```
