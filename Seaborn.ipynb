#Program to illustrate Heat Map

```python
# importing the modules
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
  
# generating 2-D 10x10 matrix of random numbers
# from 1 to 100
data = np.random.randint(low = 1,
                         high = 100,
                         size = (10, 10))
print("The data to be plotted:\n")
print(data)
  
# plotting the heatmap
hm = sn.heatmap(data = data, annot = True)
  
# displaying the plotted heatmap
plt.show()
```

#Program to illustrate Countplot

```python
# importing the required library 
import seaborn as sns
import matplotlib.pyplot as plt
 

df = sns.load_dataset('tips')
 
# count plot on single categorical variable
sns.countplot(x ='sex', data = df)
 
# Show the plot
plt.show()
```

#Program to plot vertical Bar Plot

```python
## Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
  
sns.set_context('paper')
 
# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.barplot(x = 'embark_town', y = 'age', data = titanic,
            palette = 'PuRd',ci=None 
            )
plt.legend()
plt.show()
print(titanic.columns)
```

#Program to plot Horizontal Bar Plot

```python
## Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
 
sns.set_context('paper')
# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.barplot(x = 'age', y = 'embark_town', data = titanic,
            palette = 'PuRd', orient = 'h',
            )
plt.show()
```

#Progam to illustrate  Count Plot

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
  
sns.set_context('paper')
 
# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.countplot(x = 'class', hue = 'who', data = titanic, palette = 'magma')
plt.title('Survivors')
plt.show()
```
