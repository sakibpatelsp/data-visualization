#Program to demonstrate Density plot

```python
# Density Plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kde

data = [12,33,3,4,2,1,15,6,4,13,3,13,66,44,51,34,31,2]
density = kde.gaussian_kde(data)
x = np.linspace(-2,10,300)
y=density(x)
plt.plot(x,y)
plt.title("Density Plot of the data")
plt.show()
```

#Program to find determinant of square marix

```python
#importing the scipy and numpy packages
from scipy import linalg
import numpy as np
  
#Declaring the numpy array
A = np.array([[11,12,19],[13,14,18],[17,18,14]])
#Passing the values to function
x = linalg.det(A)
  
#printing the result
print('Determinant of \n{} \n is {}'.format(A,x))
```

#Program to find eigen values of matrix

```python
#importing the scipy and numpy packages
from scipy import linalg
import numpy as np
  
#Declaring the numpy array
A = np.array([[22,11,-12],[32,1,0],[2,11,0]])
  
#Passing the values to the eig function
values, vectors = linalg.eig(A)
  
#printing the result for eigenvalues
print(values)
  
#printing the result for eigenvectors
print(vectors)
```

#Program to find eigen values and vectors of matrix

```python
#importing the scipy and numpy packages
from scipy import linalg
import numpy as np
  
#Declaring the numpy array
A = np.array([[12,4,-1],[3,0,2],[1,1,0]])
  
#Passing the values to the eig function
values, vectors = linalg.eig(A)
  
#printing the result for eigenvalues
print(values)
  
#printing the result for eigenvectors
print(vectors)
```
