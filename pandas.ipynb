```python
#Program to create a Series by passing a list of values, letting pandas create a default integer index.

import pandas as pd
import numpy as np
s = pd.Series([1, 3, 5, np.nan, 6, 8])
s
```

```python
#Program to create a DataFrame by passing a NumPy array, with a datetime index and labeled columns.

import pandas as pd
dates = pd.date_range("20130101", periods=6)
dates

df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list("ABCD"))
df
```

```python
#Program to create DataFrame by passing a dictionary of objects that can be converted into a series-like structure.
import pandas as pd
df2 = pd.DataFrame(
    {
        "A": 1.0,
        "B": pd.Timestamp("20130102"),
        "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo",
    }
)


df2
```

```python
# Program to Concatenating pandas objects together with concat()
import  pandas as pd
df = pd.DataFrame(np.random.randn(10, 4))

df

# break it into pieces
pieces = [df[:3], df[3:7], df[7:]]

pd.concat(pieces)
```
