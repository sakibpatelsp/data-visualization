# Write a Program to demonstrate Map based plot

```python
#importing Library
import plotly.express as px
#data Frame
df = px.data.gapminder().query("year == 2007")
fig = px.scatter_geo(df, locations="iso_alpha",
                     size="pop", # size of markers, "pop" is one of the columns of gapminder
                     )
fig.show()
```

#Program to Demonstrate Dendogram Plot

```python
# importing library
import plotly.figure_factory as ff

import numpy as np

X = np.random.rand(10, 12)
names = ['Omkar', 'pranav', 'Sakib', 'sharukh', 'Salman', 'aishwariya', 'katrina', 'karina', 'alia', 'Abhishek']
fig = ff.create_dendrogram(X, orientation='left', labels=names)
fig.update_layout(width=800, height=800)
fig.show()
```

#Program to demonstrate pie chart

```python
import plotly.express as px
# This dataframe has 244 lines, but 4 distinct values for `day`
df = px.data.tips()
fig = px.pie(df, values='tip', names='day')
fig.show()
```

#Progam to demonstrate line plot

```python
#importing necesary library
import plotly.express as px
import pandas as pd

df = pd.DataFrame(dict(
    x = [1, 3, 2, 4],
    y = [1, 2, 3, 4]
))
fig = px.line(df, x="x", y="y", title="Unsorted") 
fig.show()

df = df.sort_values(by="x")
fig = px.line(df, x="x", y="y", title="Sorted") 
fig.show()
```

#Program to demonstrate violin plot

```python
#impoting necessary libraries
import plotly.express as px
df = px.data.tips()
fig = px.violin(df, x="sex", y="total_bill", color="smoker")
fig.show()
```

#Program to demonstate categorical bar plot

```python
#impoting nececaary library
import plotly.express as px
df = px.data.tips()
fig = px.bar(df, x="day", y="total_bill", color="smoker", barmode="group", facet_col="sex",
             category_orders={"day": ["Thur", "Fri", "Sat", "Sun"],
                              "smoker": ["Yes", "No"],
                              "sex": ["Male", "Female"]})
fig.show()
```

#Program to demonstate categorical bar plot in ascending order

```python
#importing neccary library
import plotly.graph_objects as go
#data frame
x=['Omkar', 'Sakib', 'Sanika', 'vinayak']
fig = go.Figure(go.Bar(x=x, y=[20,13,19,14], name='Python'))
fig.add_trace(go.Bar(x=x, y=[14, 18, 19, 16], name='Java'))
fig.add_trace(go.Bar(x=x, y=[19, 17, 12, 11], name='c'))
fig.update_layout(barmode='stack')
fig.update_xaxes(categoryorder='category ascending')
fig.show()
```
